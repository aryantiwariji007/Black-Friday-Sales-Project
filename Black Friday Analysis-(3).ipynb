```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("BlackFriday.csv")
del df["Product_Category_2"]
del df["Product_Category_3"]
```

```python
df.head()
```

```python
df.groupby("Age").size().plot(kind="bar",title="Purchase Distribution by Ages")
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.countplot(x="Gender",hue="Age",data=df)
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.countplot(x="Marital_Status",hue="Age",data=df)
```

```python
sns.countplot(x="Age",hue=df["City_Category"],data=df)
```

```python
df.groupby("City_Category").size().plot(kind="pie",autopct="0.1%f")
```

```python
sns.countplot(x="Marital_Status",hue=df["City_Category"],data=df)
```

```python
sns.countplot(x="City_Category",hue="Gender",data=df)
```

```python
df.groupby("City_Category").sum()["Purchase"].plot(kind="pie",autopct="%.2f")
```

```python
sns.countplot(x="Occupation",hue="Gender",data=df)
```

```python
df.groupby("Occupation").nunique()["Product_ID"].sort_values().plot(kind='bar')
```

```python
df.groupby('Product_Category_1').size().sort_values().plot(kind='bar')
```

```python
df.groupby('Product_Category_1').sum()["Purchase"].sort_values().plot(kind='bar')
```

```python
df.groupby('Product_ID').sum()['Purchase'].nlargest(10).sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_ID').size().nlargest(10).sort_values().plot(kind = 'bar')
```

```python
sns.countplot(x="Product_Category_1",hue="Gender",data=df)
```
